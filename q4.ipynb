```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inputMatrix(n):
    x = np.empty(n)
    y = np.empty(n)
    for i in range(n):
        x[i] = int(input("X coord: "))
        y[i] = int(input("Y coord: "))
    return np.vstack((x, y, np.ones(n)))
x = inputMatrix(int(input('Enter no. of coordinates: ')))
x
```

```python
def plotFig(x):
    x_closed = np.append(x[0], x[0, 0])
    y_closed = np.append(x[1], x[1, 0])
    print(x_closed, y_closed)
    plt.plot(x_closed, y_closed)
plotFig(x)
```

```python
translation_matrix = np.eye(3)
translation_matrix[2, 0], translation_matrix[2, 1] = 1, 2
translation_matrix
x.shape
```

```python
def scaling(x, x_factor, y_factor):
    scaling_matrix = np.diag([x_factor, y_factor, 1])
    scaled_matrix = scaling_matrix @ x
    return scaled_matrix
x_factor = int(input('Scaling factor for x: '))
y_factor = int(input('scaling factor for y: '))
scaled_matrix = scaling(x, x_factor, y_factor)
plotFig(scaled_matrix)
```

```python
def translation(x, x_factor, y_factor):
    translation_matrix = np.eye(3)
    translation_matrix[0, 2], translation_matrix[1, 2] = x_factor, y_factor
    translated_matrix = translation_matrix @ x
    return translated_matrix
x_factor = int(input('Translation along x: '))
y_factor = int(input('Translation along y: '))
translated_matrix = translation(x, x_factor, y_factor)
plotFig(translated_matrix)
translated_matrix
```

```python
def rotation(x, rotation_angle):
    rotation_matrix = np.eye(3)
    rotation_matrix[0, 0] = np.cos(np.deg2rad(rotation_angle))
    rotation_matrix[0, 1] = - np.sin(np.deg2rad(rotation_angle))
    rotation_matrix[1, 0] = np.sin(np.deg2rad(rotation_angle))
    rotation_matrix[1, 1] = rotation_matrix[0, 0]
    rotated_matrix = rotation_matrix @ x
    return rotated_matrix
rotated_matrix = rotation(x, 45)
plotFig(rotated_matrix)
```

```python
def reflection(x, axis):
    reflection_matrix = np.eye(3)
    if axis == 'x':
        reflection_matrix[1, 1] = -1
    elif axis == 'y':
        reflection_matrix[0, 0] = -1
    return reflection_matrix @ x
reflected_matrix = reflection(x, input('Along axis (x or y): '))
plotFig(reflected_matrix)
```
